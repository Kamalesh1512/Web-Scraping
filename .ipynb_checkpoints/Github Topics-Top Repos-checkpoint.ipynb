### Scraping the Github Website - Topics: 
https://github.com/topics

###  Import Libraries

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
```

```python
def fetch_html_content(url):
    # header paramters
    header={"User-Agent":
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"}
    
    webpage=requests.get(url,params=header).text
    html_doc=BeautifulSoup(webpage,'html.parser')
    return html_doc
## Base url 
base_url ="https://github.com/topics"
html_doc=fetch_html_content(base_url)
```

```python
def scrape_topics(doc):
    
    topics_details=doc.find_all('div',{'class':"py-4 border-bottom d-flex flex-justify-between"})
    
    topics_dict={"Title":[],
                 "Description":[],
                 "Links":[]
                }
    base_url="https://github.com/"
    for topic in topics_details:
        #title
        topics_dict['Title'].append(topic.find('p',{"class":"f3 lh-condensed mb-0 mt-1 Link--primary"}).text)
        #description
        topics_dict['Description'].append(topic.find("p",{"class":"f5 color-fg-muted mb-0 mt-1"}).text.strip())
        #url
        topics_dict['Links'].append(base_url+topic.find("a",{'class':'no-underline flex-grow-0'})['href'])
    
    return pd.DataFrame(topics_dict)

topics_df=scrape_topics(html_doc) 
topics_df.head()
```

```python
for index,row in topics_df.iterrows():
    print(row['Title'],row['Links'])
```

```python
def scrape_repo_details(doc):
    ## repo details
    repo_details=doc.find_all("div",{"class":"d-flex flex-justify-between flex-items-start flex-wrap gap-2 my-3"})
    repo_dict={"Username":[],"Reponame":[],"Repo_Url":[],"Stars":[]}
    for repo in repo_details:
        ## Fectching Repository Details
        repo_tags=repo.find("h3",{"class":"f3 color-fg-muted text-normal lh-condensed"})
        #username ,reponame and repo url 
        username,reponame=repo_tags.find("a").text.strip(),repo_tags.find("a",{"class":"Link text-bold wb-break-word"}).text.strip()
        repo_dict['Username'].append(username)
        repo_dict['Reponame'].append(reponame)

        repo_base_url="https://github.com/"
        repo_url=repo_tags.find("a",{"class":"Link text-bold wb-break-word"})['href']
        repo_dict['Repo_Url'].append(repo_base_url+repo_url)
        ##repo rating
        stars=repo.find('span',{"id":"repo-stars-counter-star"}).text
        repo_dict['Stars'].append(stars)

    return pd.DataFrame(repo_dict)
topic_url="https://github.com//topics/3d/"
repo_html=fetch_html_content(topic_url)
repo_df=scrape_repo_details(repo_html)
```
